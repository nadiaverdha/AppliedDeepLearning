### Training Model (with fine tuning Encoder)

```python
embed_dim = 512
attention_dim = 512
decoder_dim = 512
encoder_dim = 2048
encoder_lr = 1e-4
decoder_lr = 4e-4
grad_clip = 5.
alpha_c = 1.
vocab_size = 10000
best_bleu4 = 0.
lr_decay_factor = 0.8
lr_decay_patience = 8
best_bleu4 = 0
patience = 6
model_file = '/content/drive/MyDrive/AppliedDeepLearning/image_captioning_best_5.pth'
file = '/content/drive/MyDrive/AppliedDeepLearning/model_5.csv'
num_epochs = 30

fine_tune_encoder = True
checkpoint = '/content/drive/MyDrive/AppliedDeepLearning/image_captioning_best_5.pth'
if checkpoint is None:
        decoder = DecoderRNN(attention_dim,embed_dim,decoder_dim, vocab_size, encoder_dim)
        decoder_optimizer = torch.optim.Adam(params=filter(lambda p: p.requires_grad, decoder.parameters()), lr=decoder_lr)

        encoder = EncoderCNN()
        encoder.fine_tune(fine_tune_encoder)
        encoder_optimizer = torch.optim.Adam(params=filter(lambda p: p.requires_grad, encoder.parameters()), lr=encoder_lr) if fine_tune_encoder else None

else:
      checkpoint = torch.load(model_file)
      start_epoch = checkpoint['epoch'] + 1
      best_bleu4 = checkpoint['bleu-4']
      encoder = checkpoint['encoder']
      decoder = checkpoint['decoder']
      encoder_optimizer = checkpoint['encoder_optimizer']
      decoder_optimizer = checkpoint['decoder_optimizer']

encoder = encoder.to(device)
decoder = decoder.to(device)

if fine_tune_encoder:
    encoder_lr_scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(encoder_optimizer, mode='max', factor=lr_decay_factor, patience=lr_decay_patience)
else:
        encoder_lr_scheduler  = None
decoder_lr_scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(decoder_optimizer, mode='max', factor=lr_decay_factor, patience=lr_decay_patience)
criterion = nn.CrossEntropyLoss().to(device)
```
